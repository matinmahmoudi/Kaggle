# 3. Normal Distribution

The **normal distribution**, also known as the **Gaussian distribution**, is a **continuous** probability distribution that is characterized by a **bell-shaped** curve. It is widely used in statistics and probability theory because many **real-world phenomena** tend to follow a normal distribution. The key features of the normal distribution include its **symmetric shape** and the fact that it is completely described by two parameters: the mean (μ) and the standard deviation (σ).

### Parameters

**Mean (μ):** The **center** of the distribution, representing the average or expected value.

**Standard Deviation (σ):** A measure of the spread or **dispersion** of the data. A larger standard deviation results in a wider and flatter curve, while a smaller standard deviation makes the curve narrower and taller.

### Probability Density Function (PDF)

The probability density function of the normal distribution, denoted as f(x), is given by the following formula:



The PDF provides the probability of observing a specific value x in a continuous normal distribution.

### Use Cases

**Modeling Natural Phenomena:** Normal distribution is often used to model and analyze natural phenomena, such as the heights of individuals in a population, test scores, IQ scores, and many other measurements.

**Inferential Statistics:** The normal distribution plays a crucial role in inferential statistics. Many statistical tests and methods assume that the data follows a normal distribution, which enables researchers to make inferences about populations.

**Quality Control:** In industries like manufacturing, the normal distribution is used to monitor and control the quality of products. Deviations from the expected distribution can signal quality issues.

**Financial Markets:** Returns on investments and stock prices often exhibit a distribution that is approximately normal, which is useful for risk assessment and portfolio management.

**Psychology and Social Sciences:** In psychology and social sciences, various psychological traits and behaviors are often modeled as normally distributed to understand the underlying characteristics of a population.

### Examples

**Heights of Adults:** The heights of adult humans often follow a normal distribution. The mean height might be around 170 cm (μ), and the standard deviation might be 10 cm (σ). This means that most people fall within a range close to the mean height, with fewer individuals being much taller or shorter.

**IQ Scores:** IQ scores are typically distributed with a mean of 100 (μ) and a standard deviation of 15 (σ). This distribution allows for comparisons of intelligence across a population.

**Exam Scores:** In a classroom, if exam scores follow a normal distribution with a mean score of 75 (μ) and a standard deviation of 10 (σ), you can estimate the proportion of students who scored above a certain threshold.

**Manufacturing:** In a manufacturing process, the diameter of a certain type of screw produced may follow a normal distribution with a specific mean and standard deviation. Quality control measures can be based on this distribution to detect faulty products.

The normal distribution is essential in statistics and data analysis because of its **mathematical properties** and its frequent appearance in various real-world scenarios. It **simplifies** analysis, prediction, and hypothesis testing for a wide range of applications.

### Generate Random Samples

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy.stats import t
from scipy.stats import chi2
from scipy.stats import f
```

```python
# generate samples with numpy
samples_normal_dist = np.random.normal(loc=0, scale=1, size=100)
print(samples_normal_dist)
```

### Visualize Empirical vs Theoretical distribution

## Calculate Cumulative Probability for a Specific Quantile
This script allows you to calculate the **cumulative probability** for a specific quantile in a normal distribution, and it provides both the probability of values **lower** than the quantile and the probability of values **greater** than the quantile. This is useful in statistics and **hypothesis testing** to assess the likelihood of certain events occurring in a normal distribution.  



```python
# Calculate the cumulative probability for a specific quantile
# Define the distribution parameters (mean and standard deviation for a normal distribution)
mean = 0
std_dev = 1

# Define the quantile you want to calculate the cumulative probability for
quantile = 1.96  # Replace with your desired quantile value

# Calculate the cumulative probability using the CDF
cumulative_prob = norm.cdf(quantile, loc = mean, scale = std_dev)
Probability_greater_than = 1- cumulative_prob

# Print the cumulative probability
print(f"Probability lower than {quantile}: {cumulative_prob:.2f}")
print(f"Probability greater than {quantile}: {Probability_greater_than:.2f}")
```

## Find the Quantile for a Specific Cumulative Probability 
This script allows you to find the quantile associated with a specific cumulative probability in a normal distribution, making it useful in various statistical and probability-related applications.

```python
# Find the quantile corresponding to a specific cumulative probability
# Define the distribution parameters (mean and standard deviation for a normal distribution)
mean = 0
std_dev = 1

# Define the cumulative probability for which you want to find the quantile
cumulative_prob = 0.95  # Replace with your desired cumulative probability

# Calculate the quantile using the PPF (inverse CDF)
quantile = norm.ppf(cumulative_prob, loc=mean, scale=std_dev)

# Print the quantile
print(f"Quantile for Cumulative Probability {cumulative_prob:.2f}: {quantile:.2f}")
```

## Calculate the Probability Between Quantiles
The following script is useful for determining the probability of values falling within a **specific range** defined by lower and upper quantiles in a normal distribution. It can be applied in various statistical and probability-related scenarios.



```python
# Calculate the probability between the quantiles
# Define the distribution parameters (mean and standard deviation for a normal distribution)
mean = 0
std_dev = 1

# Define the lower and upper quantiles
lower_quantile = -1.0  # Replace with your lower quantile value
upper_quantile = 1.0   # Replace with your upper quantile value

# Calculate the cumulative probabilities at the quantiles using the CDF
lower_prob = norm.cdf(lower_quantile, loc=mean, scale=std_dev)
upper_prob = norm.cdf(upper_quantile, loc=mean, scale=std_dev)

# Calculate the probability between the quantiles
probability_between_quantiles = upper_prob - lower_prob

# Print the probability
print(f"Probability between {lower_quantile:.2f} and {upper_quantile:.2f}: {probability_between_quantiles:.2f}")
```

# 4. t-Distribution (Student's t-Distribution)

The **t-distribution**, also known as Student's t-distribution, is a **continuous** probability distribution that is used in statistics to make inferences about a population. It is **similar in shape to the normal distribution** but has **heavier tails**, which makes it more suitable for handling uncertainty in small samples.

### Parameters

The t-distribution is defined by a single parameter:

**Degrees of Freedom (ν):** This parameter represents the number of observations in the sample and influences the shape of the t-distribution. As the degrees of freedom increase, the t-distribution approaches the shape of a normal distribution.

### Use Cases

**Hypothesis Testing:** It is frequently used in hypothesis testing when comparing sample means or conducting t-tests. For example, you might use the t-distribution to assess whether two sample means are significantly different.

**Estimating Confidence Intervals:** The t-distribution is used to construct confidence intervals for population parameters. When dealing with small samples, it provides a more accurate estimation of the interval width compared to the normal distribution.

**Quality Control:** In some quality control applications, especially when dealing with small sample sizes, the t-distribution is used to set tolerance limits and assess product quality.

### Examples

**Sample Mean Estimation:** If you want to estimate the mean height of a particular species of plants, you take a small sample of plants and calculate the sample mean. However, as the population standard deviation is unknown, you use the t-distribution to construct a confidence interval for the population mean.

**t-Test:** In medical research, you may conduct a t-test to compare the mean blood pressure of patients before and after a new treatment. The t-distribution helps determine whether the change in mean blood pressure is statistically significant.

**Quality Control:** In manufacturing, a company takes a small sample of products and measures a specific characteristic. The t-distribution helps in setting tolerance limits and determining whether the production process is consistent.

**Education Research:** Researchers may use the t-distribution when analyzing the effectiveness of a new teaching method. They might compare the mean test scores of students taught using the traditional method with those taught using the new method.

In summary, the t-distribution is a valuable tool in statistics for dealing with small sample sizes and making inferences when the population standard deviation is unknown. It is widely used in various fields, including science, engineering, medicine, and quality control.

# 5. Chi-Square Distribution

The **chi-square distribution** is a **continuous** probability distribution that arises in statistical inference. It is **positively skewed** and has a **single parameter**, which is the **degrees of freedom**. This distribution is frequently used in various statistical tests and estimations, particularly in the context of **hypothesis testing** and **confidence intervals**.

### Parameters

The chi-square distribution has one parameter:

**Degrees of Freedom (r):** The degrees of freedom determine the shape and characteristics of the chi-square distribution. It reflects the number of independent standard normal random variables that are squared and summed to obtain a chi-square random variable.

### Use Cases

**Hypothesis Testing:** The chi-square distribution is commonly used in hypothesis testing, especially in situations where you need to compare observed and expected frequencies, test for independence in contingency tables, or assess goodness-of-fit.

**Confidence Intervals:** It is used to construct confidence intervals for population variance when dealing with normally distributed data.

**Quality Control:** In manufacturing and quality control, chi-square tests can be employed to detect defects and evaluate product consistency.

**Medical Research:** Chi-square tests are used in medical research for analyzing the relationships between categorical variables, such as assessing the effectiveness of a treatment or examining disease risk factors.

### Examples

It describes the distribution of the ratio of two independent chi-square random variables, each divided by their respective degrees of freedom.

### Use Cases

**Analysis of Variance (ANOVA):** The F-distribution is widely used in ANOVA to compare means between two or more groups. It helps assess whether the variations between the group means are significant or if they could occur by random chance.

**Regression Analysis:** In regression analysis, the F-test is used to determine the overall significance of a regression model. It helps answer whether the model, with all its independent variables, explains a significant portion of the variation in the dependent variable.

**Quality Control:** F-distribution is used to perform variance ratio tests in quality control and manufacturing processes. It helps assess whether the variability between samples or groups is significantly different.

**Experimental Design:** In experimental design, particularly when comparing the effects of multiple factors, the F-distribution is used to assess the significance of those factors and interactions.

### Examples

**ANOVA Test:** In a clinical trial, researchers want to know if three different drug treatments have significantly different effects on patient recovery times. They use ANOVA with the F-distribution to test if there are significant differences among the treatment groups.

**Regression Analysis:** A business analyst uses linear regression to assess the impact of advertising spending and sales promotions on product sales. The F-test is used to evaluate whether the entire regression model is statistically significant.

**Quality Control:** A manufacturing company measures the diameter of bolts produced by two different machines. They use the F-distribution to compare the variances of the two samples to determine if one machine produces bolts with more consistent diameters.

**Experimental Design:** In agriculture, researchers conduct experiments to evaluate the effect of different fertilizers and irrigation methods on crop yields. They use the F-distribution to assess the significance of these factors in affecting crop yields.

In summary, the F-distribution is a fundamental tool in statistics and data analysis, particularly in hypothesis testing and regression analysis. It helps determine the significance of variation between groups or models and is essential for making informed decisions in a wide range of fields.
