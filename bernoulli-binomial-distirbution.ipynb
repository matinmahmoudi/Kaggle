## 1. Bernoulli Distribution  

The **Bernoulli distribution** is a probability distribution that describes the outcomes of a **binary** or yes-no experiment. It is named after the Swiss mathematician Daniel Bernoulli. In this distribution, there are only **two possible outcomes**, conventionally labeled as **"success"** (usually denoted as 1) and **"failure"** (usually denoted as 0). The distribution is characterized by a **single parameter**, often denoted as "**p**," which represents the probability of success.

The **probability mass function** of the Bernoulli distribution is defined as follows:

P(X = 1) = p (probability of success)

P(X = 0) = 1 - p (probability of failure)

In simple terms, if you perform a Bernoulli trial (a single experiment with two possible outcomes), the distribution tells you the likelihood of getting a success (1) or a failure (0). It's commonly used in various applications, such as modeling **coin flips** (where heads might be a success and tails a failure), **binary classification** problems in **machine learning**, and other situations where there are only two possible outcomes.

### Generate Random Samples
This code is a Python script that uses the **NumPy** library to demonstrate how to generate random samples from a Bernoulli distribution.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli
from scipy.stats import binom
```

```python
# Generate random samples from a Bernoulli distribution with p = .5
samples_bernolli = np.random.binomial(n=1, p=.5, size=500)
print(samples_bernolli)
```

### Calculate Frequency Table


```python
# Count the occurrences of each value in the samples array
counts_bernoulli = pd.Series(samples_bernolli).value_counts()

# Calculate the percentage
percentage_bernoulli = counts_bernoulli / len(samples_bernolli) * 100

# Display the counts and percentage
sample_bernoulli_df = pd.DataFrame({'Counts': counts_bernoulli, 'Percentage': percentage_bernoulli})
print(sample_bernoulli_df)
```

### Visualize Samples vs Population
This code visualizes two **bar plots** of a Bernoulli distribution with a specific probability parameter. The first plot for **samples** and other one for the **population** with a Bernoulli distribution.

```python
bern_dist = bernoulli(p=.5)
# Set the style of the plot
sns.set_theme(style="whitegrid")

# Create a figure and axes with subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the sample distribution
sns.barplot(x=sample_bernoulli_df.index, y=sample_bernoulli_df['Percentage'], ax=axs[0], palette='Blues_d')

# Customize the sample distribution plot
axs[0].set_xlabel('Outcomes')
axs[0].set_ylabel('Probability (%)')
axs[0].set_title('Sample Distribution')
axs[0].set_ylim([0, 100])  # Set the y-axis limits

# Plot the Bernoulli distribution
sns.barplot(x=[0, 1], y=bern_dist.pmf([0, 1]) * 100, ax=axs[1], palette='Greens_d')

# Customize the Bernoulli distribution plot
axs[1].set_xlabel('Outcomes')
axs[1].set_ylabel('Probability (%)')
axs[1].set_title('Bernoulli Distribution')
axs[1].set_ylim([0, 100])  # Set the y-axis limits

# Add grid lines to the plots
axs[0].grid(True, linestyle='--', linewidth=0.5)
axs[1].grid(True, linestyle='--', linewidth=0.5)

# Add a background color to the plots
axs[0].set_facecolor('#f0f0f0')
axs[1].set_facecolor('#f0f0f0')

# Add a title to the figure
fig.suptitle('Sample and Bernoulli Distributions', fontsize=14, fontweight='bold')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

## 2. Binomial Distribution

The **binomial distribution** is a probability distribution that describes **the number of successes** (usually referred to as "x") in a **fixed number of independent and identical Bernoulli trials**. In Bernoulli trials, each trial has only two possible outcomes, typically labeled as success and failure. The binomial distribution is used to calculate the probability of getting a specific number of successes in a fixed number of such trials.

Parameters:

n: The total number of trials or experiments.  

p: The probability of success in each individual trial.

The binomial distribution is commonly used in various fields like **statistics**, **quality control**, and **biology**.
It is used when you want to find the probability of achieving a specific number of successes in a fixed number of trials with binary outcomes (success or failure).
### Examples:

In a series of 10 free-throw shots (n = 10) by a basketball player, if their probability of making a basket on each shot is 0.8 (p = 0.8), you can use the binomial distribution to find the probability of making exactly 8 shots out of 10.

In quality control, you can use the binomial distribution to calculate the probability of a specific number of defective items in a sample from a production run.

In summary, the binomial distribution is a fundamental probability distribution for discrete data that helps calculate the probability of achieving a fixed number of successes in a series of binary trials, given the probability of success in each individual trial.

### Generate Random Samples
This code is a Python script that uses the **NumPy** library to demonstrate how to generate random samples from a Binomial distribution.

```python
# Generate random samples from a Binomial distribution with p = .5
samples_binomial = np.random.binomial(n=10, p=.5, size=500)
print(samples_binomial)
```

### Calculate Frequency Table
This code processes the generated samples from a Binomial distribution, calculates the **counts** and **percentages** of each unique element (0,1,...,n), and then prints the results.

```python
# Count the occurrences of each value in the samples array
counts_binomial = pd.Series(samples_binomial).value_counts()

# Calculate the percentage
percentage_binomial = counts_binomial / len(samples_binomial) * 100

# Display the counts and percentage
sample_binomial_df = pd.DataFrame({'Counts': counts_binomial, 'Percentage': percentage_binomial})
print(sample_binomial_df)
```

### Visualize Samples vs Population
This code visualizes two **bar plots** of a Binomial distribution with a specific parameters. The first plot for **samples** and other one for the **population** with a Binomial distribution.

```python
binom_dist = binom(n=10, p=.5)

# Set the style of the plot
sns.set_theme(style="whitegrid")

# Create a figure and axes with subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the sample distribution
sns.barplot(x=sample_binomial_df.index, y=sample_binomial_df['Percentage'], ax=axs[0], palette='Blues_d')

# Customize the sample distribution plot
axs[0].set_xlabel('Outcomes')
axs[0].set_ylabel('Probability (%)')
axs[0].set_title('Sample Distribution')
axs[0].set_ylim([0, 100])  # Set the y-axis limits

# Plot the Bernoulli distribution
sns.barplot(x=np.arange(11), y=binom_dist.pmf(np.arange(11)) * 100, ax=axs[1], palette='Greens_d')
# Customize the Bernoulli distribution plot
axs[1].set_xlabel('Outcomes')
axs[1].set_ylabel('Probability (%)')
axs[1].set_title('Binomial Distribution')
axs[1].set_ylim([0, 100])  # Set the y-axis limits

# Add grid lines to the plots
axs[0].grid(True, linestyle='--', linewidth=0.5)
axs[1].grid(True, linestyle='--', linewidth=0.5)

# Add a background color to the plots
axs[0].set_facecolor('#f0f0f0')
axs[1].set_facecolor('#f0f0f0')

# Add a title to the figure
fig.suptitle('Sample and Binomial Distributions', fontsize=14, fontweight='bold')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

The following script calculates and displays the **PMF** of a binomial distribution for a given set of values, and it also plots and shows the **CDF** of the same distribution. It's a simple but useful script for visualizing and understanding the behavior of a binomial random variable.

```python
x = np.arange(11)  # Define the range of values for x
pmf = binom_dist.pmf(x)  # Calculate the PMF for each value of x
cdf = binom_dist.cdf(x)  # Calculate the CDF for each value of x

# Print the PMF and CDF
for i in range(len(x)):
    print(f'P(X = {x[i]}) = {pmf[i]}, F(X <= {x[i]}) = {cdf[i]}')
```

```python
sns.set_theme(style="whitegrid")

# Plot the CDF of the binomial distribution
plt.plot(x, cdf * 100, color='blue', linewidth=2, marker='o', markersize=5)

# Add labels and title
plt.xlabel('Outcomes')
plt.ylabel('Cumulative Probability (%)')
plt.title('CDF of Binomial Distribution')

# Add grid lines
plt.grid(True, linestyle='--', linewidth=0.5)


# Show the plot
plt.show()
```
