# Poisson Distribution - Discrete

The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space when these events happen with a known constant mean rate and independently of the time since the last event.

## Key Characteristics

- **Parameter λ (Lambda):** This is the average number of events per interval. It’s the only parameter of the Poisson distribution.
- **Support:** The possible values of the Poisson distribution are non-negative integers (0, 1, 2, ...).

## Applications in Data Science

The Poisson distribution is particularly useful in data science for modeling the frequency of events happening within a fixed period of time when these events are independent. For example, it can model the number of users visiting a website in a minute or the number of system errors occurring in a day.

## Example: Website Traffic Analysis

Imagine you are analyzing the traffic to a popular data science blog (for example medium), and you find that the site receives, on average, 30 visitors per hour. We can model the number of visitors in any given hour using the Poisson distribution.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
lambda_ = 30  # Average number of events per interval

# Generate Poisson data
data_poisson = np.random.poisson(lambda_, 1000)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_poisson, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)  

# Add a vertical line for the mean
plt.axvline(x=lambda_, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Improve the title and axis labels
plt.title('Poisson Distribution - Website Traffic per Hour', fontsize=16)
plt.xlabel('Number of Visitors', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Binomial Distribution - Discrete

The Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials of a binary experiment. It is defined by two parameters: the number of trials $(n)$ and the probability of success $(p)$ in each trial.

## Key Characteristics

- **Parameters:**
  - $(n)$ (Number of Trials): The number of times the experiment is performed.
  - $(p)$ (Probability of Success): The probability of success on an individual trial.
- **Support:** The possible values of the Binomial distribution are integers ranging from 0 to $(n)$.

## Applications in Data Science

The Binomial distribution is widely used in data science for modeling binary outcomes. For example, it can be used to model the number of clicks on an advertisement out of several views or the number of defective items in a batch of produced goods.

## Example: Click-Through Rate Analysis

Suppose you want to analyze the effectiveness of a new banner ad, knowing that historically, the click-through rate (CTR) is about 20%. If the ad is shown to 100 visitors, the number of clicks can be modeled using the Binomial distribution.

## Comparison with Poisson Distribution

- **Similarities:** Both are discrete distributions used to model the number of occurrences of events.
- **Differences:**
  - The Poisson distribution is used when the events are independent and occur with a known constant mean rate, typically over time or space.
  - The Binomial distribution applies when the outcomes are binary and derived from a fixed number of trials, each with the same probability of success.

The Poisson distribution is the limit of the Binomial distribution for large $(n)$ and small $(p)$, where the mean $ (\lambda)$ is given by:

$$
\lambda = n \times p
$$

## Normal Approximation

When $(n)$ is large and $(p)$ is near 0.5, the Binomial distribution approximates the Normal distribution. This is due to the Central Limit Theorem, which states that the distribution of the sample mean will be approximately normal if the sample size is sufficiently large.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
n = 100  # Number of trials
p = 0.2  # Probability of success

# Generate Binomial data
data_binomial = np.random.binomial(n, p, 1000)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_binomial, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)  # Label for histogram set to None

# Add a vertical line for the mean
mean = n * p
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Improve the title and axis labels
plt.title('Binomial Distribution - Probability of Clicks on Ad', fontsize=16)
plt.xlabel('Number of Clicks', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Geometric Distribution - Discrete

The Geometric distribution is a discrete probability distribution that models the number of trials needed to achieve the first success in a sequence of Bernoulli trials (trials with two possible outcomes, typically success or failure). The probability of success on each trial is constant.

## Key Characteristics

- **Parameter p (Probability of Success):** The probability of success on an individual trial.
- **Support:** The possible values of the Geometric distribution are positive integers (1, 2, 3, ...).

## Applications in Data Science

The Geometric distribution is useful in data science for modeling the number of attempts required until a success is achieved, such as the number of page loads until a user clicks on an advertisement.

## Example: User Engagement Analysis

Consider a scenario where a user must pass through multiple tutorial screens before finding the 'Start' button. If the probability of passing each screen on the first attempt is 0.2, you can model the number of screens a user will see before success using the Geometric distribution.



## Explanation of Decreasing Probabilities in Geometric Distribution

The Geometric distribution is used to model the number of trials needed to achieve the first success in a series of independent Bernoulli trials. Here's why the probabilities decrease as the number of trials increases:

- **Exponential Decay:** In this distribution, the probability of obtaining the first success on the $k$-th trial is given by $P(X = k) = (1-p)^{k-1} \cdot p$. Here, $p$ is the probability of success on a single trial, and $(1-p)^{k-1}$ is the probability of failing the first $(k-1)$ trials. This factor decreases exponentially with each additional trial, explaining why the likelihood of a first success declines as the number of trials increases.
- **Increasing Unlikelihood of Consecutive Failures:** The probability decreases because, with each additional trial, having a sequence of consecutive failures becomes increasingly unlikely. Since each trial is independent and has a constant probability of success, the chance of continuously failing diminishes the likelihood of first achieving success as more trials are conducted.

This characteristic of the Geometric distribution shows that successes are statistically more likely to occur earlier in the trial process, and the probability of first achieving success decreases with each subsequent trial.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameter
p = 0.2  # Probability of success

# Generate Geometric data
data_geometric = np.random.geometric(p, 1000)

# Calculate the mean
mean = 1 / p

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_geometric, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add a vertical line for the mean and update the label to include its value
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {mean:.2f}')

# Improve the title and axis labels
plt.title('Geometric Distribution - Probability of Trials Until First Success', fontsize=16)
plt.xlabel('Number of Trials', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Negative Binomial Distribution - Discrete

The Negative Binomial distribution is a discrete probability distribution that models the number of trials required to achieve a predetermined number of successes in a sequence of independent Bernoulli trials. This distribution is a generalization of the Geometric distribution, which is the special case of the Negative Binomial distribution when the number of successes required is one.

## Key Characteristics

- **Parameters:**
  - \( r \) (Number of Successes): The target number of successes that must be achieved.
  - \( p \) (Probability of Success): The probability of success on an individual trial.
- **Support:** The possible values of the Negative Binomial distribution are integers starting from \( r \) (since achieving \( r \) successes requires at least \( r \) trials).

## Applications in Data Science

The Negative Binomial distribution is commonly used in scenarios where the event success rate is low or highly variable, such as the number of trials needed to achieve a certain number of conversions in online advertising, or modeling over-dispersed count data in ecological studies.

## Example: Customer Conversion Analysis

Imagine a scenario where a new tech product requires at least 5 trials (demos or sessions) to convert a customer, with each trial having a success probability of 0.2. The distribution of the number of trials to achieve 5 conversions can be modeled using the Negative Binomial distribution.


## Behavior of the Negative Binomial Distribution

The Negative Binomial distribution often first increases and then decreases due to the balance between achieving a required number of successes and the increasing improbability of continued trials. Initially, as the number of trials increases, it becomes more likely to reach the required number of successes. However, once this peak probability is passed, it becomes increasingly unlikely to continue needing more trials for the same number of successes, leading to a decrease in probability as the number of trials continues to grow. This distribution highlights the transition from an insufficient to a sufficient number of trials needed to achieve the set number of successes.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
r = 5  # Number of successes
p = 0.2  # Probability of success

# Generate Negative Binomial data
data_neg_binomial = np.random.negative_binomial(r, p, 1000)

# Calculate the mean
mean = r / p

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_neg_binomial, color='deepskyblue', binwidth=2, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add a vertical line for the mean and update the label to include its value
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {mean:.2f}')

# Improve the title and axis labels
plt.title('Negative Binomial Distribution - Trials Needed for Multiple Successes', fontsize=16)
plt.xlabel('Number of Trials', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Hypergeometric Distribution - Discrete

The Hypergeometric distribution models the number of successes in a sample drawn without replacement from a finite population containing a fixed number of successes and failures. It is used when the probability of success changes on each draw because the composition of the population changes after each draw.

## Key Characteristics

- **Parameters:**
  - \( N \) (Population Size): The total number of items in the population.
  - \( K \) (Number of Successes in Population): The total number of items categorized as successes in the population.
  - \( n \) (Sample Size): The number of items to draw from the population.
- **Support:** The possible values range from the maximum of 0 or \( n + K - N \) to the minimum of \( n \) and \( K \).

## Applications in Data Science

The Hypergeometric distribution is useful in scenarios where the population is not large enough to assume independence between draws, such as quality control, lot testing, or when sampling without replacement from a small population batch.

## Example: Quality Control Testing

Consider a batch of 100 items, where 20 are defective. If you randomly select 10 items for testing, the distribution of the number of defective items in your sample can be modeled using the Hypergeometric distribution.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
N = 100  # Population size
K = 20   # Number of successes in population
n = 10   # Sample size

# Generate Hypergeometric data
data_hypergeom = np.random.hypergeometric(K, N-K, n, 1000)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_hypergeom, color='deepskyblue', binwidth=1, edgecolor='navy',
                  linewidth=1.5, stat="probability")

# Improve the title and axis labels
plt.title('Hypergeometric Distribution - Number of Defective Items Found', fontsize=16)
plt.xlabel('Number of Defective Items', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```

# Multinomial Distribution - Discrete

The Multinomial distribution extends the Binomial distribution to experiments where each trial can result in one of multiple possible outcomes. Each outcome has a fixed probability.

## Key Characteristics

- **Parameters:**
  - **n (Number of Trials):** The total number of trials.
  - **p (Probability of Outcomes):** A list of probabilities for each outcome, with the sum equalling 1.
- **Support:** The result is a vector of non-negative integers that sum to n, with each entry representing the count of a particular outcome.

## Applications in Data Science

The Multinomial distribution is essential for analyzing categorical data in scenarios like election predictions, consumer choice behavior, and classification tasks in machine learning.

## Example: Election Voting Scenario

Consider an election with three candidates. Suppose 100 people vote, with the following probabilities for each candidate:
- Candidate A: 40%
- Candidate B: 35%
- Candidate C: 25%

We can model the distribution of votes using the Multinomial distribution to predict possible outcomes.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
n = 100  # Number of trials (voters)
probabilities = [0.4, 0.35, 0.25]  # Probabilities for candidates A, B, and C

# Generate Multinomial data
data_multinomial = np.random.multinomial(n, probabilities, size=1)

# Set the dark style
sns.set_style("darkgrid")

# Visualization
fig, ax = plt.subplots()
categories = ['Candidate A', 'Candidate B', 'Candidate C']
colors = ['skyblue', 'steelblue', 'navy']  # Shades of blue for each candidate
ax.bar(categories, data_multinomial[0], color=colors)

# Improve the title and axis labels
plt.title('Multinomial Distribution - Election Voting Results', fontsize=16)
plt.xlabel('Candidates', fontsize=14)
plt.ylabel('Number of Votes', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

# Exponential Distribution - Continuous

The Exponential distribution is a continuous probability distribution used to model the time or space between events in a Poisson process. It describes situations where independent events occur at a constant rate.

## Key Characteristics

- **Parameter \( \lambda \) (Rate):** This is the rate at which events occur. Higher \( \lambda \) values lead to shorter times between events.
- **Support:** The distribution is defined for all non-negative real numbers. This means the values of \( x \), which represent time or space between events, must be zero or positive (x >= 0).

## Applications in Data Science

The Exponential distribution is crucial in data science for modeling 'time to event' data, such as:
- The time until a machine fails.
- The time until the next customer arrives at a service station.

## Example: Website Traffic Analysis

Consider modeling the time between successive visits to a website, where visits occur at an average rate of 3 visits per hour. This scenario can be represented by an Exponential distribution.


### Why Does Probability Decrease Over Time in the Exponential Distribution?

The exponential distribution describes the time between events in a Poisson process. This distribution is characterized by a constant rate (lambda) at which events occur. As time passes, the probability of an event occurring in a short interval decreases. This is because the exponential distribution models the "memoryless" property, where the probability of an event occurring in the next interval is independent of how much time has already passed. The probability density function (PDF) of the exponential distribution is given by:

f(t) = lambda * exp(-lambda * t)

Here, lambda is the rate parameter, and t is the time. As t increases, exp(-lambda * t) decreases, showing that the probability decreases over time.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameter for Website Traffic Analysis
lambda_rate = 3  # Average number of visits per hour (rate of visits)

# Generate Exponential data representing the time between visits
data_exponential = np.random.exponential(scale=1/lambda_rate, size=1000)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_exponential, kde=True, color='deepskyblue', binwidth=0.1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None, kde_kws={'bw_adjust': 0.3})

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the mean time between visits
mean = 1 / lambda_rate
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean: {:.2f} hours'.format(mean))

# Improve the title and axis labels
plt.title('Exponential Distribution - Time Between Website Visits', fontsize=16)
plt.xlabel('Time Between Visits (hours)', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Uniform Distribution - Continuous

The Uniform distribution is a continuous probability distribution that is characterized by equal probability for all outcomes within a specific interval. It is used for modeling scenarios where each outcome within a certain range is equally likely.

## Key Characteristics

- **Parameters (a, b):** These define the lower bound (a) and upper bound (b) of the distribution, with all values within this range being equally probable.
- **Support:** The distribution covers values from the minimum \(a\) to the maximum \(b\), inclusively.

## Applications in Data Science

The Uniform distribution is particularly useful for simulations and modeling equal likelihood scenarios. It is commonly applied in:
- Simulating response times of servers where each response time within a given range is equally probable.
- Random sampling within a fixed interval for Monte Carlo simulations.

## Example: Server Response Times

Consider a scenario where a network of servers is designed to respond to requests within a uniform time range from 100 to 200 milliseconds. This model can be used to simulate and analyze the consistency and reliability of server response times across the network.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters for Uniform Distribution
min_response_time = 100  # Minimum response time in milliseconds
max_response_time = 200  # Maximum response time in milliseconds

# Calculate the mean for the uniform distribution
mean_response_time = (min_response_time + max_response_time) / 2

# Generate Uniform distribution data representing server response times
data_uniform = np.random.uniform(low=min_response_time, high=max_response_time, size=1000)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_uniform, kde=True, color='deepskyblue', binwidth=3, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the mean response time
plt.axvline(x=mean_response_time, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {mean_response_time:.0f} ms')

# Improve the title and axis labels
plt.title('Uniform Distribution - Server Response Times', fontsize=16)
plt.xlabel('Response Time (milliseconds)', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Normal Distribution - Continuous

The Normal or Gaussian distribution is one of the most important distributions in statistics due to its frequent occurrence in the natural and social sciences. It is characterized by its bell-shaped curve, which is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean.

## Key Characteristics

- **Parameters (μ, σ):** The mean (μ) and standard deviation (σ) are the primary parameters. The mean indicates the central value, and the standard deviation denotes the spread of the distribution.
- **Support:** The distribution covers all real numbers, extending from negative infinity to positive infinity.

## Applications in Data Science

The Normal distribution is essential for many statistical methods and tests, including hypothesis testing and confidence intervals, and is used to model:
- Errors or residuals in regression analysis.
- Biological characteristics like heights or weights.
- Performance measurements in areas such as education or sports.

## Example: Adult Male Heights

Assuming the heights of adult males in a certain population follow a Normal distribution, with a mean height of 175 cm and a standard deviation of 10 cm, we can use this model to perform various statistical analyses, such as calculating the probability of a randomly selected individual being within a certain height range.


#### Key Characteristics

1. **Symmetry:**
   - The normal distribution is perfectly symmetrical around its mean. This means that the left half of the distribution is a mirror image of the right half.

2. **Bell-shaped Curve:**
   - The normal distribution has a bell-shaped curve, with the highest point at the mean (μ).

3. **Mean, Median, and Mode:**
   - In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.

4. **Standard Deviation (σ):**
   - The standard deviation determines the width of the bell curve. A smaller standard deviation results in a narrower curve, while a larger standard deviation results in a wider curve.

5. **68-95-99.7 Rule:**
   - This rule states that in a normal distribution:
     - About 68% of the data falls within one standard deviation (μ ± σ) of the mean.
     - About 95% of the data falls within two standard deviations (μ ± 2σ) of the mean.
     - About 99.7% of the data falls within three standard deviations (μ ± 3σ) of the mean.


### Standardization of the Normal Distribution

Standardization is the process of rescaling a normal distribution to have a mean of zero and a standard deviation of one. This is also known as converting to the standard normal distribution or applying the Z-transform.

#### Formula for Standardization

To standardize a data point \( x \) from a normal distribution with mean \( \mu \) and standard deviation \( \sigma \), use the following formula:

    z = (x - μ) / σ

Where:
- \( x \) is the original data point.
- \( μ \) is the mean of the dataset.
- \( σ \) is the standard deviation of the dataset.
- \( z \) is the standardized value.

#### Key Characteristics

1. **Mean and Standard Deviation:**
   - After standardization, the transformed data will have a mean of 0 and a standard deviation of 1.

2. **Comparability:**
   - Standardized values (z-scores) allow for comparing data points from different normal distributions on a common scale.

3. **Standard Normal Distribution:**
   - The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters for Normal Distribution
mean_height = 175  # Mean height in cm
std_dev = 10  # Standard deviation in cm

# Generate Normal distribution data for heights
data_normal = np.random.normal(loc=mean_height, scale=std_dev, size=1000)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_normal, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the mean height
plt.axvline(x=mean_height, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {mean_height} cm')

# Improve the title and axis labels
plt.title('Normal Distribution - Adult Male Heights', fontsize=16)
plt.xlabel('Height (cm)', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Log-normal Distribution - Continuous

The Log-normal distribution is used to describe data that is positively skewed, with a long tail to the right. It arises when a random variable's logarithm is normally distributed, making it ideal for representing variables that are greater than zero.

## Key Characteristics

- **Parameters (μ, σ):** These are the mean and standard deviation of the variable's logarithm, not the variable itself. The original variable's distribution becomes skewed due to the exponential transformation.
- **Support:** The distribution is defined for all positive real numbers, as the actual values are the exponent of the normal distribution.

## Applications in Data Science

The Log-normal distribution is particularly useful for modeling:
- Economic data such as incomes, where large values are possible but small values are more common.
- Sizes of living organisms in biology.
- The time required to complete tasks where there is a minimum time but no sharp upper limit.

## Example: Income Distribution

Assuming that the income distribution in a certain region follows a Log-normal distribution, with a mean log-income of 4 (in natural log scale) and a standard deviation of 1.5, this can be used to analyze economic inequality, predict economic trends, or establish income thresholds for policy making.


#### Key Characteristics

1. **Positive Values:**
   - The log-normal distribution is defined only for positive real numbers. This makes it suitable for modeling distributions where values cannot be negative, such as income, prices, or stock prices.

2. **Right Skewed:**
   - The log-normal distribution is right-skewed, meaning it has a long tail on the right side. This property is useful for modeling data that exhibit this kind of skewness.

3. **Parameters:**
   - The log-normal distribution is characterized by two parameters:
     - **Location (μ):** The mean of the logarithm of the variable.
     - **Scale (σ):** The standard deviation of the logarithm of the variable.

4. **Relationship to Normal Distribution:**
   - If \(X\) is normally distributed with mean \(μ\) and standard deviation \(σ\), then \(Y = e^X\) follows a log-normal distribution. Conversely, if \(Y\) is log-normally distributed, then \(\log(Y)\) is normally distributed.

#### Why Use the Log-Normal Distribution?

1. **Modeling Skewed Data:**
   - The log-normal distribution is ideal for modeling data that are positively skewed and cannot be negative. This includes many real-world phenomena such as income distribution, stock prices, and certain types of biological data.

2. **Multiplicative Processes:**
   - It is suitable for modeling processes where the underlying factors multiply rather than add. For example, in finance, the returns of an investment are often multiplicative, making the log-normal distribution a good fit for modeling stock prices.

3. **Natural Phenomena:**
   - Many natural phenomena, such as the size of particles or the growth rates of organisms, follow a log-normal distribution due to the multiplicative effects of many small independent factors.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Adjusted Parameters for Log-normal Distribution
mu = 2  # Mean of the logarithm of the variable (log-income)
sigma = 0.75  # Standard deviation of the logarithm of the variable

# Generate Log-normal distribution data for income
data_log_normal = np.random.lognormal(mean=mu, sigma=sigma, size=1000)

# Calculate the mean of the actual data
actual_mean = np.exp(mu + sigma**2 / 2)

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_log_normal, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the actual mean of the log-normal distribution
plt.axvline(x=actual_mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {actual_mean:.2f}')

# Improve the title and axis labels
plt.title('Log-normal Distribution - Income Distribution', fontsize=16)
plt.xlabel('Income', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Gamma Distribution - Continuous

The Gamma distribution is used to describe the sum of exponentially distributed random variables. It is particularly useful for modeling waiting times or the total time of consecutive events in many fields of engineering, meteorology, and health sciences.

## Key Characteristics

- **Parameters (shape, scale):** The shape parameter \(k\) and the scale parameter \( \theta \) determine the distribution's form. \(k\) is often the number of events, and \( \theta \) is the average time between events.
- **Support:** The distribution is defined for all positive real numbers, as it represents times or other positive measurements.

## Applications in Data Science

The Gamma distribution is widely used in:
- Modeling the time until death in medical research.
- Estimating the amount of rainfall accumulated in a reservoir over a period.
- Queueing theory to model the total service time for multiple consecutive services.

## Example: Service Time at a Bank

Assuming the service time for each customer at a bank follows an exponential distribution with a mean of 10 minutes, we can use the Gamma distribution to model the total service time for 5 consecutive customers. This can help in resource planning and optimizing service processes.

# Beta Distribution - Continuous

The Beta distribution is a versatile and flexible distribution used to model variables that are constrained within an interval, typically [0, 1]. It is particularly useful for representing proportions and probabilities where outcomes are percentages or fractions.

## Key Characteristics

- **Parameters (α, β):** These shape parameters, alpha (α) and beta (β), determine the distribution's skewness and how the values are concentrated. Higher values of α or β indicate more concentration around the high or low ends, respectively.
- **Support:** The distribution is defined between 0 and 1 inclusive, which is ideal for modeling percentages or probability measures.

## Applications in Data Science

The Beta distribution is extensively used in:
- Bayesian statistics for modeling prior distributions over probabilities.
- Quality control for modeling the fraction of defectives in a manufacturing process.
- Finance to model the risk of investments when the outcomes are proportions.

## Example: Machine Uptime Analysis

Assuming the operational uptime of a machine over a year is measured, and the proportion of uptime is expected to be around 90% with slight variability, the Beta distribution can model this uncertainty. Let's consider α = 9 and β = 1 to reflect a high likelihood of the machine being operational most of the time.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters for Beta Distribution
alpha = 9  # Alpha parameter
beta = 1   # Beta parameter

# Generate Beta distribution data for machine uptime
data_beta = np.random.beta(alpha, beta, 1000)

# Calculate the mean of the distribution
actual_mean = alpha / (alpha + beta)

# Set the dark style
sns.set_style("darkgrid")


# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_beta, kde=True, color='deepskyblue', binwidth=0.01, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the actual mean of the Beta distribution
plt.axvline(x=actual_mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {actual_mean:.2f}')

# Improve the title and axis labels
plt.title('Beta Distribution - Machine Uptime Analysis', fontsize=16)
plt.xlabel('Proportion of Uptime', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Chi-squared Distribution - Continuous

The Chi-squared distribution is a fundamental tool in statistical inference, primarily used in hypothesis testing and in estimating variances. It is derived from the sum of the squares of a number of independent standard normal random variables.

## Key Characteristics

- **Degrees of Freedom (k):** The key parameter of the Chi-squared distribution. It is equal to the number of independent standard normal random variables squared and summed to form the Chi-squared statistic.
- **Support:** The distribution is defined for all positive real numbers.

## Applications in Data Science

The Chi-squared distribution is widely used in:
- Goodness-of-fit testing to compare observed data against a theoretical distribution.
- Tests of independence in contingency tables to analyze the relationship between categorical variables.
- Confidence interval estimation for population variance in normal distributions.

## Example: Goodness-of-Fit Testing

In an experiment to determine whether a die is fair, the number of times each face appears after several rolls can be compared against the expected counts using a Chi-squared test. If the die has six sides, we use a Chi-squared distribution with 5 degrees of freedom (one less than the number of categories) to determine the p-value of observing the given outcomes if the die were fair.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Degrees of Freedom
df = 5  # For a fair six-sided die

# Generate Chi-squared distribution data
data_chi_squared = np.random.chisquare(df, 1000)

# Calculate the mean of the distribution
actual_mean = df

# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_chi_squared, kde=True, color='deepskyblue', binwidth=.5, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the actual mean of the Chi-squared distribution
plt.axvline(x=actual_mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {actual_mean:.2f}')

# Improve the title and axis labels
plt.title('Chi-squared Distribution - Goodness-of-Fit Testing', fontsize=16)
plt.xlabel('Chi-squared Value', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Student's t-Distribution - Continuous

The Student's t-distribution is a key tool in statistics, primarily used when estimating the mean of a normally distributed population when the sample size is small and the population standard deviation is unknown. It resembles the normal distribution but has heavier tails, providing a more reliable estimate in small samples.

## Key Characteristics

- **Degrees of Freedom (ν):** This parameter is typically one less than the sample size and indicates the level of estimation uncertainty in the tails of the distribution.
- **Support:** The distribution covers all real numbers, much like the normal distribution, but with a wider spread to accommodate more uncertainty.

## Applications in Data Science

Student's t-distribution is essential for:
- Calculating confidence intervals and conducting hypothesis tests for small sample sizes.
- Analyzing investment returns, particularly when sample sizes are small and data variability is high.
- Quality assurance testing in manufacturing when only a few product samples are available.

In a manufacturing process, if a quality inspector wishes to estimate the average thickness of a batch of glass panels using a small sample of panels, the Student's t-distribution can be used. # Fisher’s F-Distribution - Continuous

Fisher’s F-distribution is widely used in statistical analysis, particularly in the context of hypothesis testing involving variances. It arises when comparing the ratios of variances between two normally distributed populations, which is common in the analysis of variance (ANOVA) tests.

## Key Characteristics

- **Degrees of Freedom (df1, df2):** The degrees of freedom for the numerator (df1) and the denominator (df2) are determined by the sample sizes of the two groups being compared, minus one for each group.
- **Support:** The distribution is defined for all positive real numbers, as it represents the ratio of two variances.

## Applications in Data Science

Fisher's F-distribution is essential for:
- Conducting ANOVA tests to determine if there are significant differences in means across multiple groups.
- Comparing variances in a two-sample F-test to assess equality of variances between two populations.
- Evaluating the proportion of variance explained in regression analysis through the F-test for overall significance.

## Example: Clinical Trial Variance Analysis

In a clinical trial comparing two treatments, researchers might collect data on the response times of patients. To determine if the variability in response times differs significantly between the treatments, an F-test can be employed. For example, if treatment A has 15 patients and treatment B has 20, the degrees of freedom for the numerator would be 14 (df1=15-1) and for the denominator would be 19 (df2=20-1).


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Degrees of Freedom for two samples
df1 = 14  # df for treatment A
df2 = 19  # df for treatment B

# Generate Fisher's F-distribution data
data_fisher_f = np.random.f(df1, df2, 1000)

# Calculate the mean of the distribution (only defined for df2 > 2)
if df2 > 2:
    actual_mean = df2 / (df2 - 2)
else:
    actual_mean = 'undefined'
    
    
# Set the dark style
sns.set_style("darkgrid")

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_fisher_f, kde=True, color='deepskyblue', binwidth=0.1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.5, label='KDE')

# Add a vertical line for the actual mean of the Fisher's F-distribution
if df2 > 2:
    plt.axvline(x=actual_mean, color='darkred', linestyle='--', linewidth=2, label=f'Mean: {actual_mean:.2f}')

# Improve the title and axis labels
plt.title('Fisher’s F-Distribution - Clinical Trial Variance Analysis', fontsize=16)
plt.xlabel('F Value', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Central Limit Theorem

The Central Limit Theorem (CLT) is a fundamental principle in statistics that states the distribution of the sum (or average) of a large number of independent, identically distributed variables approaches a normal distribution, regardless of the original distribution.

This means that if you take sufficiently large samples from any distribution, the distribution of the sample means will tend to be normal.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to demonstrate the Central Limit Theorem
def central_limit_theorem_demo(sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.exponential(scale=2, size=sample_size)
        means.append(np.mean(sample))
    return means

# Parameters
sample_sizes = [10, 30, 100, 300]
num_samples = 1000

# Plotting the distributions of sample means for different sample sizes
plt.figure(figsize=(12, 8))
for size in sample_sizes:
    sample_means = central_limit_theorem_demo(size, num_samples)
    sns.histplot(sample_means, kde=True, label=f'Sample size: {size}', stat='density')

plt.title('Central Limit Theorem Demonstration')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.show()
```
