# Poisson Distribution - Discrete

The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space when these events happen with a known constant mean rate and independently of the time since the last event.

## Key Characteristics

- **Parameter Î» (Lambda):** This is the average number of events per interval. Itâ€™s the only parameter of the Poisson distribution.
- **Support:** The possible values of the Poisson distribution are non-negative integers (0, 1, 2, ...).

## Applications in Data Science

The Poisson distribution is particularly useful in data science for modeling the frequency of events happening within a fixed period of time when these events are independent. For example, it can model the number of users visiting a website in a minute or the number of system errors occurring in a day.

## Example: Website Traffic Analysis

Imagine you are analyzing the traffic to a popular data science blog, and you find that the site receives, on average, 30 visitors per hour. We can model the number of visitors in any given hour using the Poisson distribution.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
lambda_ = 30  # Average number of events per interval

# Generate Poisson data
data_poisson = np.random.poisson(lambda_, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_poisson, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)  # Label for histogram set to None

# Add a vertical line for the mean
plt.axvline(x=lambda_, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Adding shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Improve the title and axis labels
plt.title('Poisson Distribution - Probability of Events per Hour', fontsize=16)
plt.xlabel('Number of Events', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Binomial Distribution - Discrete

The Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials of a binary experiment. It is defined by two parameters: the number of trials \(n\) and the probability of success \(p\) in each trial.

## Key Characteristics

- **Parameters:**
  - \( n \) (Number of Trials): The number of times the experiment is performed.
  - \( p \) (Probability of Success): The probability of success on an individual trial.
- **Support:** The possible values of the Binomial distribution are integers ranging from 0 to \( n \).

## Applications in Data Science

The Binomial distribution is widely used in data science for modeling binary outcomes. For example, it can be used to model the number of clicks on an advertisement out of several views or the number of defective items in a batch of produced goods.

## Example: Click-Through Rate Analysis

Suppose you want to analyze the effectiveness of a new banner ad, knowing that historically, the click-through rate (CTR) is about 20%. If the ad is shown to 100 visitors, the number of clicks can be modeled using the Binomial distribution.

## Comparison with Poisson Distribution

- **Similarities:** Both are discrete distributions used to model the number of occurrences of events.
- **Differences:**
  - The Poisson distribution is used when the events are independent and occur with a known constant mean rate, typically over time or space.
  - The Binomial distribution applies when the outcomes are binary and derived from a fixed number of trials, each with the same probability of success.
  
### The Poisson distribution is the limit of the binomial distribution for large N.



```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
n = 100  # Number of trials
p = 0.2  # Probability of success

# Generate Binomial data
data_binomial = np.random.binomial(n, p, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_binomial, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)  # Label for histogram set to None

# Add a vertical line for the mean
mean = n * p
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Adding shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Improve the title and axis labels
plt.title('Binomial Distribution - Probability of Clicks on Ad', fontsize=16)
plt.xlabel('Number of Clicks', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
