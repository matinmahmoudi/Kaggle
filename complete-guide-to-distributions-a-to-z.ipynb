# Poisson Distribution - Discrete

The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space when these events happen with a known constant mean rate and independently of the time since the last event.

## Key Characteristics

- **Parameter Î» (Lambda):** This is the average number of events per interval. Itâ€™s the only parameter of the Poisson distribution.
- **Support:** The possible values of the Poisson distribution are non-negative integers (0, 1, 2, ...).

## Applications in Data Science

The Poisson distribution is particularly useful in data science for modeling the frequency of events happening within a fixed period of time when these events are independent. For example, it can model the number of users visiting a website in a minute or the number of system errors occurring in a day.

## Example: Website Traffic Analysis

Imagine you are analyzing the traffic to a popular data science blog, and you find that the site receives, on average, 30 visitors per hour. We can model the number of visitors in any given hour using the Poisson distribution.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
lambda_ = 30  # Average number of events per interval

# Generate Poisson data
data_poisson = np.random.poisson(lambda_, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_poisson, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)  # Label for histogram set to None

# Add a vertical line for the mean
plt.axvline(x=lambda_, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Adding shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Improve the title and axis labels
plt.title('Poisson Distribution - Probability of Events per Hour', fontsize=16)
plt.xlabel('Number of Events', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Binomial Distribution - Discrete

The Binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials of a binary experiment. It is defined by two parameters: the number of trials \(n\) and the probability of success \(p\) in each trial.

## Key Characteristics

- **Parameters:**
  - \( n \) (Number of Trials): The number of times the experiment is performed.
  - \( p \) (Probability of Success): The probability of success on an individual trial.
- **Support:** The possible values of the Binomial distribution are integers ranging from 0 to \( n \).

## Applications in Data Science

The Binomial distribution is widely used in data science for modeling binary outcomes. For example, it can be used to model the number of clicks on an advertisement out of several views or the number of defective items in a batch of produced goods.

## Example: Click-Through Rate Analysis

Suppose you want to analyze the effectiveness of a new banner ad, knowing that historically, the click-through rate (CTR) is about 20%. If the ad is shown to 100 visitors, the number of clicks can be modeled using the Binomial distribution.

## Comparison with Poisson Distribution

- **Similarities:** Both are discrete distributions used to model the number of occurrences of events.
- **Differences:**
  - The Poisson distribution is used when the events are independent and occur with a known constant mean rate, typically over time or space.
  - The Binomial distribution applies when the outcomes are binary and derived from a fixed number of trials, each with the same probability of success.
  
### The Poisson distribution is the limit of the binomial distribution for large N.



```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
n = 100  # Number of trials
p = 0.2  # Probability of success

# Generate Binomial data
data_binomial = np.random.binomial(n, p, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_binomial, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)  # Label for histogram set to None

# Add a vertical line for the mean
mean = n * p
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Adding shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Improve the title and axis labels
plt.title('Binomial Distribution - Probability of Clicks on Ad', fontsize=16)
plt.xlabel('Number of Clicks', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Geometric Distribution - Discrete

The Geometric distribution is a discrete probability distribution that models the number of trials needed to achieve the first success in a sequence of Bernoulli trials (trials with two possible outcomes, typically success or failure). The probability of success on each trial is constant.

## Key Characteristics

- **Parameter p (Probability of Success):** The probability of success on an individual trial.
- **Support:** The possible values of the Geometric distribution are positive integers (1, 2, 3, ...).

## Applications in Data Science

The Geometric distribution is useful in data science for modeling the number of attempts required until a success is achieved, such as the number of page loads until a user clicks on an advertisement.

## Example: User Engagement Analysis

Consider a scenario where a user must pass through multiple tutorial screens before finding the 'Start' button. If the probability of passing each screen on the first attempt is 0.2, you can model the number of screens a user will see before success using the Geometric distribution.



## Explanation of Decreasing Probabilities in Geometric Distribution

The Geometric distribution models the number of trials until the first success in repeated Bernoulli trials. The key characteristics that lead to decreasing probabilities are:

- **Exponential Decay:** The probability of achieving the first success on the \( k \)-th trial is given by \( P(X = k) = (1-p)^{k-1} \times p \), where \( p \) is the probability of success on each trial. The term \( (1-p)^{k-1} \) represents the probability of failing \( k-1 \) times in a row, which exponentially decreases as \( k \) increases.
- **Increasing Unlikelihood:** As the number of trials increases without a success, the likelihood of just then achieving success becomes progressively lower. This reflects the natural expectation that a sequence of consecutive failures becomes less probable if each trial has a fixed chance of success.

This distribution emphasizes that early successes are more probable, and the chance of first encountering success grows slimmer with each additional trial.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameter
p = 0.2  # Probability of success

# Generate Geometric data
data_geometric = np.random.geometric(p, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_geometric, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Add a vertical line for the mean
mean = 1 / p
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Improve the title and axis labels
plt.title('Geometric Distribution - Probability of Trials Until First Success', fontsize=16)
plt.xlabel('Number of Trials', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Negative Binomial Distribution - Discrete

The Negative Binomial distribution is a discrete probability distribution that models the number of trials required to achieve a predetermined number of successes in a sequence of independent Bernoulli trials. This distribution is a generalization of the Geometric distribution, which is the special case of the Negative Binomial distribution when the number of successes required is one.

## Key Characteristics

- **Parameters:**
  - \( r \) (Number of Successes): The target number of successes that must be achieved.
  - \( p \) (Probability of Success): The probability of success on an individual trial.
- **Support:** The possible values of the Negative Binomial distribution are integers starting from \( r \) (since achieving \( r \) successes requires at least \( r \) trials).

## Applications in Data Science

The Negative Binomial distribution is commonly used in scenarios where the event success rate is low or highly variable, such as the number of trials needed to achieve a certain number of conversions in online advertising, or modeling over-dispersed count data in ecological studies.

## Example: Customer Conversion Analysis

Imagine a scenario where a new tech product requires at least 5 trials (demos or sessions) to convert a customer, with each trial having a success probability of 0.2. The distribution of the number of trials to achieve 5 conversions can be modeled using the Negative Binomial distribution.


## Behavior of the Negative Binomial Distribution

The Negative Binomial distribution often first increases and then decreases due to the balance between achieving a required number of successes and the increasing improbability of continued trials. Initially, as the number of trials increases, it becomes more likely to reach the required number of successes. However, once this peak probability is passed, it becomes increasingly unlikely to continue needing more trials for the same number of successes, leading to a decrease in probability as the number of trials continues to grow. This distribution highlights the transition from an insufficient to a sufficient number of trials needed to achieve the set number of successes.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
r = 5  # Number of successes
p = 0.2  # Probability of success

# Generate Negative Binomial data
data_neg_binomial = np.random.negative_binomial(r, p, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_neg_binomial, kde=True, color='deepskyblue', binwidth=1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Add a vertical line for the mean
mean = r / p
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean')

# Improve the title and axis labels
plt.title('Negative Binomial Distribution - Trials Needed for Multiple Successes', fontsize=16)
plt.xlabel('Number of Trials', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

# Hypergeometric Distribution - Discrete

The Hypergeometric distribution models the number of successes in a sample drawn without replacement from a finite population containing a fixed number of successes and failures. It is used when the probability of success changes on each draw because the composition of the population changes after each draw.

## Key Characteristics

- **Parameters:**
  - \( N \) (Population Size): The total number of items in the population.
  - \( K \) (Number of Successes in Population): The total number of items categorized as successes in the population.
  - \( n \) (Sample Size): The number of items to draw from the population.
- **Support:** The possible values range from the maximum of 0 or \( n + K - N \) to the minimum of \( n \) and \( K \).

## Applications in Data Science

The Hypergeometric distribution is useful in scenarios where the population is not large enough to assume independence between draws, such as quality control, lot testing, or when sampling without replacement from a small population batch.

## Example: Quality Control Testing

Consider a batch of 100 items, where 20 are defective. If you randomly select 10 items for testing, the distribution of the number of defective items in your sample can be modeled using the Hypergeometric distribution.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
N = 100  # Population size
K = 20   # Number of successes in population
n = 10   # Sample size

# Generate Hypergeometric data
data_hypergeom = np.random.hypergeometric(K, N-K, n, 1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_hypergeom, kde=False, color='deepskyblue', binwidth=1, edgecolor='navy',
                  linewidth=1.5, stat="probability")

# Improve the title and axis labels
plt.title('Hypergeometric Distribution - Number of Defective Items Found', fontsize=16)
plt.xlabel('Number of Defective Items', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```

# Multinomial Distribution - Discrete

The Multinomial distribution extends the Binomial distribution to experiments where each trial can result in one of multiple possible outcomes. Each outcome has a fixed probability.

## Key Characteristics

- **Parameters:**
  - **n (Number of Trials):** The total number of trials.
  - **p (Probability of Outcomes):** A list of probabilities for each outcome, with the sum equalling 1.
- **Support:** The result is a vector of non-negative integers that sum to n, with each entry representing the count of a particular outcome.

## Applications in Data Science

The Multinomial distribution is essential for analyzing categorical data in scenarios like election predictions, consumer choice behavior, and classification tasks in machine learning.

## Example: Election Voting Scenario

Consider an election with three candidates. Suppose 100 people vote, with the following probabilities for each candidate:
- Candidate A: 40%
- Candidate B: 35%
- Candidate C: 25%

We can model the distribution of votes using the Multinomial distribution to predict possible outcomes.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameters
n = 100  # Number of trials (voters)
probabilities = [0.4, 0.35, 0.25]  # Probabilities for candidates A, B, and C

# Generate Multinomial data
data_multinomial = np.random.multinomial(n, probabilities, size=1)

# Visualization
fig, ax = plt.subplots()
categories = ['Candidate A', 'Candidate B', 'Candidate C']
colors = ['skyblue', 'steelblue', 'navy']  # Shades of blue for each candidate
ax.bar(categories, data_multinomial[0], color=colors)

# Improve the title and axis labels
plt.title('Multinomial Distribution - Election Voting Results', fontsize=16)
plt.xlabel('Candidates', fontsize=14)
plt.ylabel('Number of Votes', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

# Exponential Distribution - Continuous

The Exponential distribution is a continuous probability distribution used to model the time or space between events in a Poisson process. It describes situations where independent events occur at a constant rate.

## Key Characteristics

- **Parameter \( \lambda \) (Rate):** This is the rate at which events occur. Higher \( \lambda \) values lead to shorter times between events.
- **Support:** The distribution is defined for all non-negative real numbers. This means the values of \( x \), which represent time or space between events, must be zero or positive (x >= 0).

## Applications in Data Science

The Exponential distribution is crucial in data science for modeling 'time to event' data, such as:
- The time until a machine fails.
- The time until the next customer arrives at a service station.

## Example: Website Traffic Analysis

Consider modeling the time between successive visits to a website, where visits occur at an average rate of 3 visits per hour. This scenario can be represented by an Exponential distribution.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# Parameter for Website Traffic Analysis
lambda_rate = 3  # Average number of visits per hour (rate of visits)

# Generate Exponential data representing the time between visits
data_exponential = np.random.exponential(scale=1/lambda_rate, size=1000)

# Set up the figure and axes for a cleaner layout
plt.figure(figsize=(12, 7))
ax = sns.histplot(data_exponential, kde=True, color='deepskyblue', binwidth=0.1, edgecolor='navy', 
                  linewidth=1.5, stat="probability", label=None)

# Add shading under the curve
kde_x, kde_y = ax.lines[0].get_data()
ax.fill_between(kde_x, kde_y, color="skyblue", alpha=0.2, label='KDE')

# Add a vertical line for the mean time between visits
mean = 1 / lambda_rate
plt.axvline(x=mean, color='darkred', linestyle='--', linewidth=2, label='Mean: {:.2f} hours'.format(mean))

# Improve the title and axis labels
plt.title('Exponential Distribution - Time Between Website Visits', fontsize=16)
plt.xlabel('Time Between Visits (hours)', fontsize=14)
plt.ylabel('Probability', fontsize=14)

# Customize the ticks for better readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show probability with two decimal places
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

# Optional: Add grid lines for better visual comparison
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Enhance with a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```
