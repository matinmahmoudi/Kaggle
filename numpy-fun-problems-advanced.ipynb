# Q1 - Population Verification Problem

https://www.practiceprobs.com/problemsets/python-numpy/advanced/population-verification/

You manage a local department for the Census responsible for measuring the population of each block in your city . Even though you could do it yourself, for each of the last five years, youâ€™ve tasked this job to your subordinate, Jim.

Each year, Jim gives you a 2x4 array of his population estimates where each element of the array represents a city block. After five years, you have a 5x2x4 array of population estimates called jim where (i,j,k) represents Jimâ€™s population estimate for block (j,k) of year i.



```python
import numpy as np

generator = np.random.default_rng(2357)
jim = np.round(generator.normal(loc=100, scale=5, size=(5,2,4)))

print(jim)
```

You donâ€™t fully trust Jimâ€™s estimates because you see him spending an ungodly amount of time on Facebook. So, each year, you go behind his back and measure the population of two city blocks. After five years, you have the following data:

blocks, a 5x2x2 array indicating which blocks you measured each year where (i,j) gives the coordinates for the jth block you measured in year i

pops, a corresponding 5x2 array where (i,j) gives the population you measured for the jth block in year i

```python
blocks = np.array([
    [[0,2],[1,3]],
    [[1,2],[0,0]],
    [[0,0],[1,2]],
    [[1,1],[0,3]],
    [[0,1],[1,0]]
])

pops = np.array([
    [100, 105],
    [110, 92],
    [95, 99],
    [89, 107],
    [101, 98]
])
```

How many times was Jimâ€™s estimate at least 10% higher or lower than your estimate?


## Solution

```python
# Index Jim's estimates using the blocks array
jim_estimates = jim[np.arange(5)[:, None], blocks[:, :, 0], blocks[:, :, 1]]
```

```python
# Calculate the absolute percentage difference between Jim's estimate and your estimate
diff = np.abs(jim_estimates - pops) / pops
```

```python
# Count the number of times the absolute percentage difference is greater than or equal to 10%
count = np.sum(diff >= 0.1)
```

# Q2 - Prime Locations Problem

https://www.practiceprobs.com/problemsets/python-numpy/advanced/prime-locations/

Given a 10x10x10 array of zeros, set (i,j,k) = 1 if i is odd, j is even, and k is prime.

In other words, set these elements to 1: (1,0,2), (1,0,3), (1,0,5), (1,0,7), (1,2,2), ...

```python
import numpy as np

chewy = np.zeros((10,10,10))

print(chewy)
```

## Solution 1

```python
# Define a function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(np.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# Set the elements of chewy to 1 if i is odd, j is even, and k is prime
for i in range(10):
    for j in range(10):
        for k in range(10):
            if i % 2 == 1 and j % 2 == 0 and is_prime(k):
                chewy[i, j, k] = 1

chewy
```

## Solution 2 - more professional

```python
i = np.array([1,3,5,7,9])
j = np.array([0,2,4,6,8])
k = np.array([2,3,5,7])
chewy[i[:, None, None], j[None, :, None], k[None, None, :]] = 1

chewy
```

# Q3 - The Game Of Doors Problem

https://www.practiceprobs.com/problemsets/python-numpy/advanced/the-game-of-doors/

Youâ€™re preparing for a game show where you play the following game.

Youâ€™re faced with four doors    
Behind one door is cash money ðŸ¤‘
Behind the other three doors is steaming pile of broccoli ðŸ¥¦

If you pick the door with money, you get the option to play again at the risk of losing the money youâ€™ve collected thus far. If you keep winning, you can play a maximum of three rounds before the game show host kicks you off his show.

You have some strategy ideas youâ€™d like to evaluate before you go on the show. You hire an unpaid intern to scour through historic recordings of the show and log data on where prizes were located and the value of each prize.

Your intern gets back to you with the following matrices:

prize_doors: Here, element (i,j) identifies the door hiding the prize in the ith game of round j
prizes: Here (i, j) gives the prize amount in the ith game of round j


```python
import numpy as np

prize_doors = np.array([
    [1, 0, 2],
    [0, 0, 1],
    [3, 3, 1],
    [1, 2, 0],
    [2, 1, 1]
])

prizes = np.array([
    [100, 150, 500],
    [200, 300, 250],
    [150, 100, 325],
    [425, 200, 100],
    [200, 250, 300]
])
```

Build a 5x3x4 matrix where (i,j,k) represents the prize behind door k of round j of game i.



## Solution 1

```python
prize_matrix = np.zeros((5, 3, 4))

for i in range(5):
    for j in range(3):
        for k in range(4):
            door = prize_doors[i, j]
            prize = prizes[i, j]
            if door == k:
                prize_matrix[i, j, k] = prize

prize_matrix
```

## Solution 2 - more professional

```python
answer = np.zeros((5, 3, 4))
answer[np.arange(5)[:, None], np.arange(3)[None, :], prize_doors] = prizes

answer
```

# Q5 - Freckle Problem

https://www.practiceprobs.com/problemsets/python-numpy/advanced/freckle/

Given a 1-D array called freckle,

```python
import numpy as np
freckle = np.array([2,7,3,1,0,5])
```

perform the following sequence of operations.
1. Sort the elements in descending order	7, 5, 3, 2, 1, 0
2. Take their cumulative sum	7, 12, 15, 17, 18, 18
3. Undo the sort from step 1	17, 7, 15, 18, 18, 12

## Solution

```python
# sort descending
sorted_idxs = np.argsort(freckle)[::-1]
freckle_descending = freckle[sorted_idxs]
freckle_descending
```

```python
# cumulative sum
freckle_cumsum = np.cumsum(freckle_descending)
freckle_cumsum
```

```python
# undo sort
undo_sorted_idxs = np.argsort(sorted_idxs)
result = freckle_cumsum[undo_sorted_idxs]
result
```

# Q6 - Get Rich Bot Problem

https://www.practiceprobs.com/problemsets/python-numpy/advanced/get-rich-bot/

You've developed an algorithmic stock trading bot to strike it rich ðŸ¤‘, but you want to backtest it before putting it into production. You have price data on 10 stocks, each tracked at 8 regular time intervals.


```python
import numpy as np

rng = np.random.default_rng(1234)
prices = [rng.normal(rng.lognormal(3, 1), size=8).round(2) for i in range(10)]
prices = np.vstack(prices)

print(prices)
```

For each stock, your algorithm suggests the best time to buy the stock and the best time to sell the stock.

```python
trade_idxs = [np.sort(rng.choice(8, size=2, replace=False)) for i in range(10)]
trade_idxs = np.vstack(trade_idxs)

print(trade_idxs)
```

trade_idxs identifies the index of the buy price and sell price for each stock. So, the holding periods look like this.



```python
"""
| trade_idxs |     0 |     1 |     2 |      3 |     4 |     5 |     6 |     7 |
|:-----------|------:|------:|------:|-------:|------:|------:|------:|------:|
| [0 1]      |  4.1  |  4.78 | ----- |  ----- | ----- | ----- | ----- | ----- |
| [3 6]      | ----- | ----- | ----- |  28.84 | 27.06 | 26.17 | 28.76 | ----- |
| [2 4]      | ----- | ----- | 34.56 |  34.98 | 32.63 | ----- | ----- | ----- |
| [1 5]      | ----- | 19.73 | 21.68 |  20.61 | 19.87 | 17.86 | ----- | ----- |
| [0 3]      |  4.24 |  6.39 |  5.32 |   3.87 | ----- | ----- | ----- | ----- |
| [2 7]      | ----- | ----- | 99.02 |  99.94 | 98.28 | 99.15 | 98.19 | 98.49 |
| [2 3]      | ----- | ----- | 09.01 | 105.19 | ----- | ----- | ----- | ----- |
| [1 3]      | ----- | 16.34 | 15.88 |  17.53 | ----- | ----- | ----- | ----- |
| [3 6]      | ----- | ----- | ----- |  94.48 | 95.42 | 93.75 | 93.54 | ----- |
| [1 4]      | ----- | 17.43 | 16.53 |  20.2  | 18.68 | ----- | ----- | ----- |
"""
```

Given prices and trade_idxs, calculate the average price of each holding period.

## Solution 1

```python
average_prices = np.array([prices[i, trade_idxs[i, 0]:trade_idxs[i, 1]+1].mean() for i in range(prices.shape[0])])
average_prices
```

## Solution 2

```python
# Identify which elements to "mask"
col_idxs = np.arange(prices.shape[1])
mask = (col_idxs < trade_idxs[:, [0]]) | (col_idxs > trade_idxs[:, [1]])
```

```python
# convert mask elements to nan
prices[mask] = np.nan
```

```python
# Calculate the row means
np.nanmean(prices, axis=1)
```
